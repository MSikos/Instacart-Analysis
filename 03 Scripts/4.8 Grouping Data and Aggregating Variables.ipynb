# 5. Create loyalty flag for customers using transform() and loc()

```python
# Use groupby and transform() functions with np.max to determine maximum order number for each user
ords_prods_merged_updated['max_order'] = ords_prods_merged_updated.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Validate the function updated our dataframe
ords_prods_merged_updated.head(20)
```

```python
# Use loc() function to make loyalty flag column with the given criteria of > 40 orders is a loyal customer
ords_prods_merged_updated.loc[ords_prods_merged_updated['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Use loc() function to make loyalty flag column with the given criteria of between 10 and 40 orders is a regular customer
ords_prods_merged_updated.loc[(ords_prods_merged_updated['max_order'] <= 40) & (ords_prods_merged_updated['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Use loc() function to make loyalty flag column with the given criteria of < 10 orders is a new customer
ords_prods_merged_updated.loc[ords_prods_merged_updated['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Validate new column data and frequency of values
ords_prods_merged_updated['loyalty_flag'].value_counts(dropna = False)
```

```python
# Validate updated dataframe headings
ords_prods_merged_updated.head()
```

# 6. Investigate spending habits of customer groups

```python
# Use groupby and agg() functions to determine basic statistics of spending habits by loyalty group
ords_prods_merged_updated.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

According to the data, new customers spend approximately 3 dollars more on average then a loyal customer.

# 7. Create a spending flag for each user based on the average price across all of their orders

```python
# Use the transform() function to determine the mean price of products purchased by each user
ords_prods_merged_updated['mean_product_price'] = ords_prods_merged_updated.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Validate the new column was created successfully and only view it and user_id
ords_prods_merged_updated[['user_id', 'mean_product_price']].head(20)
```

```python
# Create spending_flag column using loc() with criteria of 10 or more being a high spender
ords_prods_merged_updated.loc[ords_prods_merged_updated['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Create spending_flag column using loc() with criteria of lass than 10 being a low spender
ords_prods_merged_updated.loc[ords_prods_merged_updated['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Validate new column created successfully and check frequency of new column
ords_prods_merged_updated['spending_flag'].value_counts(dropna = False)
```

```python
# View column headers of only the user_id, mean_relevant_price, and spending_flag columns
ords_prods_merged_updated[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

# 8. Determine frequent versus non-frequent customers

```python
# Use groupby and transform() functions to determine median number of days between orders for each user
ords_prods_merged_updated['median_days_between_orders'] = ords_prods_merged_updated.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Validate new column created successfully and view column headers for user_id, days_since_last_order, and new column
ords_prods_merged_updated[['user_id', 'days_since_last_order', 'median_days_between_orders']].head(20)
```

```python
# Create order frequency flag using loc() function based on the criteria if median_days_between_orders is > 20 then the user is a non-frequent customer
ords_prods_merged_updated.loc[ords_prods_merged_updated['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# Create order frequency flag using loc() function based on the criteria if median_days_between_orders is between 10 and 20 then the user is a regular customer
ords_prods_merged_updated.loc[(ords_prods_merged_updated['median_days_between_orders'] <= 20) & (ords_prods_merged_updated['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
# Create order frequency flag using loc() function based on the criteria if median_days_between_orders is less than 10 then the user is a frequent customer
ords_prods_merged_updated.loc[ords_prods_merged_updated['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Validate new column was created successfully and check value counts of new order_frequency column
ords_prods_merged_updated['order_frequency'].value_counts(dropna = False)
```

There are rows that have not been labeled after our functions. This could be customers with only one order with Instacart and no other orders, which would be reflected in the not a number seen in the order_frequency values above.
