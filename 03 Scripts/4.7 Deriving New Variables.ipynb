## 5. Defining a function

```python
# Define a function using el-else logic

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function to the dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Validate the function values
df['price_range'].value_counts(dropna = False)
```

```python
# Check what the max value in the price column is
df['prices'].max()
```

```python
# Create the same function using .loc() statement
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Create the mid-range condition using .loc()
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Create low-range condition using .loc()
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Validate conditions using .loc() to ensure each row applied
df['price_range_loc'].value_counts(dropna = False)
```

## 7. Determining which days of the week do most orders take place

```python
# Which days of the week are the busiest (Saturday = 0 increasing with Friday = 6)
ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

```python
# Create a busiest day of the week column (busiest day) that shows the busiest day, the least busy day, and regular days (average busy)
result = []

for value in ords_prods_merged["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check the result of the function
result
```

```python
# Adding the new function to the ords_prods_merged dataframe by creating a new column called busiest day
ords_prods_merged['busiest_day'] = result
```

```python
# Validate the results of the new column
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## 3. Identify the busiest hours of the day and label the time frames as Most orders, average orders, or fewest orders
